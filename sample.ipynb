```python
from pitchiq_preprocess import PreprocessData
from pitchiq_train import ModelTrainer
```

```python
processor = PreprocessData(config_file="config.json")
processor.fetch_data()
df = processor.preprocess()
data_path = processor.save(out_dir="pitcher_data")
print(f"data processed, saved to {data_path}")
```

```python
mt = ModelTrainer(df=df)
mt.train_test_split()
mt.train_model()
mt.evaluate_model()
```
